This notebook provides a high-level overview of participant spending patterns, support categories, and potential cost anomalies based on simulated NDIS-style data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('simulated_ndis_data.csv')
df.head()
```

## Basic Statistics

```python
df.describe(include='all')
```

## Distribution of Monthly Spend

```python
plt.figure(figsize=(10,6))
sns.histplot(df['monthly_spend'], kde=True, bins=40)
plt.title('Distribution of Monthly Spend')
plt.xlabel('Monthly Spend ($AUD)')
plt.ylabel('Frequency')
plt.show()
```

## Spend by Support Category

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='support_category', y='monthly_spend', data=df)
plt.title('Monthly Spend by Support Category')
plt.xlabel('Support Category')
plt.ylabel('Monthly Spend ($AUD)')
plt.show()
```
